```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
def growth_model( y,t,B,I):
    return -B*I*y
```

```python
t = np.linspace(0, 10, 100)
y0 = 10
B = 0.9
I = 1

y = odeint(growth_model,y0,t,args=(B, I))
```

```python
plt.figure(figsize=(7, 4))
plt.plot(t, y, 'C1--', lw=2)
plt.plot(t, y *2, 'C2--', lw=2)
plt.xlabel('t')
plt.ylabel('S')
plt.title(' model epidemiologiczny – SIR')
plt.xlim(0, 10)
plt.ylim(0, 11)
plt.grid(linestyle='--')
```
