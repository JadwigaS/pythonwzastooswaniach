```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
def S_model( S,t,B,I):
    return -B*I*S
def R_model(I,t,y):
    return y*I
def I_model(S,t,B,I,y):
    return B*I*S-y*I


```

```python
def diff(N,t,params):
    S,R,I = N
    B,y = params
    derivs=[S_model(S,t,B,I),R_model(I,t,y),I_model(S,t,B,I,y)]
    return derivs

```

```python
t = np.linspace(0, 30, 300)
B=0.0001
y=0.1
S0=10000
R0=0
I0=1
params=[B,y]
N0=[S0,R0,I0]


solved= odeint(diff,N0,t,args=(params,))
```

```python
fig=plt.figure(1,figsize=(7, 4))
ax1=fig.add_subplot(311)
ax1.plot(t,solved[:,0])
ax1.set_ylabel('S')
ax1=fig.add_subplot(312)
ax1.plot(t,solved[:,1])
ax1.set_ylabel('R')
ax1=fig.add_subplot(313)
ax1.plot(t,solved[:,2])
ax1.set_ylabel('I')
plt.tight_layout()
plt.show
#plt.plot(t, solved, 'C1--', lw=2)


#plt.title(' model epidemiologiczny â€“ SIR')
#plt.xlim(0, 10)
#plt.ylim(0, 11)
#plt.grid(linestyle='--')
```
