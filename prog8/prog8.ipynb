```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

```python
x =sp.Function('x')
w = sp.Symbol('omega')
y=sp.Symbol('gamma')
t=sp.Symbol('t')
x0=sp.Symbol('x_0')
v0=sp.Symbol('v_0')
eq = sp.Eq(x(t).diff(t,t)+x(t).diff(t)*y+x(t)*w**2, 0)
eq
```

```python
sol=sp.dsolve(eq, x(t),ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol
```

```python
s1=sol.subs({x0: 1, v0: -1, w: 2,y:2})
s1
s2=sol.subs({x0: 1, v0: -1, w: 2,y:-0.5})
```

```python
sp.plot(s1.rhs, (t, 0, 10))
sp.plot(s2.rhs, (t, 0, 10))
```
